# EDA 

```python
plt.figure(figsize = (20, 15))
plotnumber = 1

for column in num_cols:
    if plotnumber <= 14:
        ax = plt.subplot(3, 5, plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column)
        
    plotnumber += 1

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (20, 15))
plotnumber = 1

for column in cat_cols:
    if plotnumber <= 14:
        ax = plt.subplot(3, 5, plotnumber)
        sns.countplot(data[column] ,palette = 'rocket')
        plt.xlabel(column)
        
    plotnumber += 1

plt.tight_layout()
plt.show()
```

## 2-Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cat_cols:
    data[col] = le.fit_transform(data[col])
```

```python
data
```

```python
for columnName in cat_cols:
  data[columnName].fillna(data[columnName].mode()[0], inplace=True)
  
for columnName in num_cols:
  data[columnName].fillna(data[columnName].mean(), inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data = data.drop(columns='id')
```

```python
data
```

```python
x = data.drop('classification' ,axis=1)  
y = data['classification']
```

```python
print("Class Distribution before Balancing:")
print(y.value_counts())
```

```python
sns.countplot(x='classification', data=data)
plt.title('Class Distribution Before Balancing')
plt.show()
```

# 3-Balancing Data

# a-UnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler
RUS = RandomUnderSampler(random_state=45)
x_res, y_res = RUS.fit_resample(x, y)
```

```python
print("Class Distribution After Under-sampling:")
print(y_res.value_counts())
```

```python
data_resampled = pd.concat([x_res, y_res], axis=1)
sns.countplot(x='classification', data=data_resampled)
plt.title('Class Distribution After Under-sampling')
plt.show()
```

# b-OverSmapling

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_res, y_res = smote.fit_resample(x, y)
```

```python
print("Class Distribution After Over-sampling:")
print(y_res.value_counts())
```

```python
data_resampled = pd.concat([x_res, y_res], axis=1)
sns.countplot(x='classification', data=data_resampled)
plt.title('Class Distribution After Over-sampling')
plt.show()
```

# 3-Feature Selection

```python
target = 'classification'
features = ['age', 'sg', 'pc', 'hemo', 'sod', 'pcv']
correlations = data[features + [target]].corr()[target].drop(target)
print("Feature Correlations with Target:")
print(correlations)
```

```python
plt.figure(figsize=(10, 6))
correlations.sort_values().plot(kind='barh')
plt.title('Feature Correlations with Target')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.show()
```

# 4-Apply algorithm

# A-RandomForest [features]

```python
from sklearn.model_selection import train_test_split
X = data[features]
Y = data[target]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(criterion = 'entropy', n_estimators = 30,random_state=50)
RFC.fit(X_train, Y_train)

Y_pred_RFC = RFC.predict(X_test)
```

# A-Analysis Results

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
RFC_ACC=accuracy_score(Y_test, Y_pred_RFC)
RFC_ACC = accuracy_score(Y_test, Y_pred_RFC)
RFC_PREC = precision_score(Y_test, Y_pred_RFC)
RFC_REC = recall_score(Y_test, Y_pred_RFC)
RFC_F1 = f1_score(Y_test, Y_pred_RFC)

print(f"Accuracy: {RFC_ACC:.3f}")
print('********************')
print(f"Precision: {RFC_PREC:.3f}")
print('********************')
print(f"F1 Score: {RFC_F1:.3f}")
print('********************')
print(f"Recall: {RFC_REC:.3f}")
print('********************')
```

```python
from sklearn.metrics import confusion_matrix

confusion_RFC = confusion_matrix(Y_test, Y_pred_RFC)
print(f"Confusion Matrix:\n ")

print('True Negative:',confusion_RFC[0,0])
print('False Negative:',confusion_RFC[1,0])
print('False positive:',confusion_RFC[0,1])
print('True pasitive:',confusion_RFC[1,1])

plt.figure(figsize=(3,3))
plt.title('Confusion Matrix')
plt.figure
sns.heatmap(confusion_RFC,annot=True , linewidth=0.5) 
 
```
